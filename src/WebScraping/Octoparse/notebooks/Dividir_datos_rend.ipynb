### Notebook final de preparación para Octoparse

En este nb necesitamos dividir las urls entre porteros y jugadores, dado que los datos estadísticos ofrecidos por la página transfermarkt son distintos para las distintas posiciones y no es posible realizar una extracción automática de los datos conjuntos

```python
import pandas as pd
import openpyxl
```

```python
df_bajadas = pd.read_excel('../data/Urls_Rend_Bajadas.xlsx', index_col=0)
df_subidas = pd.read_excel('../data/Urls_Rend_Subidas.xlsx', index_col=0)
```

```python
df_completo_subidas = pd.read_excel('../data/ValoresDeMercado_Subidas.xlsx', index_col=0)
df_completo_bajadas = pd.read_excel('../data/ValoresDeMercado_Bajadas.xlsx', index_col=0)
```

## Unimos subidas y bajadas

```python
unir_df_1 = pd.concat([df_bajadas,df_subidas],axis=0).drop_duplicates().reset_index(drop=True)
unir_df_1
```

```python
unir_df_2 = pd.concat([df_completo_bajadas,df_completo_subidas],axis=0).drop_duplicates().reset_index(drop=True)
unir_df_2
```

#### Normalizamos el texto para poder casar el mayor número de jugadores posible

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
        ("à", "a"),
        ("è", "e"),
        ("ì", "i"),
        ("ò", "o"),
        ("ù", "u"),
        ("ä", "a"),
        ("ë", "e"),
        ("ï", "i"),
        ("ö", "o"),
        ("ü", "u"),
        ('ã', 'a'),
        ('ć', 'c'),
        ('ñ', 'n'),
        ('ê', 'e'),
        ('î', 'i'),
        ('ø', 'o'),
        ('â', 'a'),
        ('%c4%87','c'),
        ('ç','c'),
        ('ô','o'),
        ('-', ' ')
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s

print(normalize("¡Hólá, múndó!"))
print(normalize("¡HÓLÁ, MÚNDÓ!"))
```

```python
import string
```

```python
lista_nombres_df1 =[]
lista_nombres_df2 =[]
for i in unir_df_1.Nombre:
    lista_nombres_df1.append(string.capwords(normalize(i)))
for j in unir_df_2.Nombre:
    lista_nombres_df2.append(string.capwords(normalize(j)))

unir_df_1.Nombre = lista_nombres_df1
unir_df_2.Nombre = lista_nombres_df2
```

## Unión completa

```python
df_final = unir_df_1.merge(unir_df_2, how='inner',on='Nombre').drop_duplicates('Nombre').dropna().reset_index(drop=True)
df_final
```

### Dividimos entre porteros y jugadores para sacar los datos estadísticos

```python
df_porteros = df_final[df_final['Posición'] == 'Portero'].reset_index(drop=True)
df_porteros
```

```python
df_jugadores =  df_final[df_final['Posición'] != 'Portero'].reset_index(drop=True)
df_jugadores
```

```python
df_porteros.to_excel('../data/Urls_porteros.xlsx')
df_jugadores.to_excel('../data/Urls_jugadores.xlsx')
```
