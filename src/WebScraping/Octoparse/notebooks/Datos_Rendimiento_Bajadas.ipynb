### Notebook de preparación para el uso de Octoparse

En este nb, utilizaremos BeautifulSoup para sacar las Urls donde se muestra el Rendimiento por Competición de cada jugador

```python
#Servicio y driver de Chrome de Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

#Las opciones que vamos a tener para buscar elementos
from selenium.webdriver.common.by import By

#Para cuando queramos mandar pulsaciones de teclado
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.common.action_chains import ActionChains

from bs4 import BeautifulSoup as bs
import requests
import datetime as dt
import time
import numpy as np
import pandas as pd
import openpyxl
import string
```

```python
df_links = pd.read_excel('../data/ValoresDeMercado_Bajadas.xlsx', index_col=0)
df_links.iloc[0,-1]
```

```python
url_principal = 'https://www.transfermarkt.es'
```

##### Creamos el bucle para extraer las urls

```python
lista_urls = []
contador = 1
headers = {"User-Agent":"Mozilla/5.0"}
for url in df_links.iloc[2749:,-1]:
    response = requests.get(url, headers=headers)
    soup = bs(response.text, 'lxml')

    print('Jugador' + str(contador))
    contador+=1

    rend_por_competicion = soup.find_all('a')
    for i in rend_por_competicion:
        if 'Datos de rendimiento detalla' in i.get_text():
            headers = {"User-Agent":"Mozilla/5.0"}
            response_detallados = requests.get(url_principal + i['href'], headers=headers)
            soup_detallados = bs(response_detallados.content, 'lxml')

            url_rend_liga = soup_detallados.find_all('div',class_='box')
            for i in url_rend_liga:
                if 'competi' in i.get_text():
                    lista_urls.append(url_principal + i.find_all('a')[-1]['href'])



```

### Sacamos los nombres de los jugadores de la propia url

```python
nombres = []
for nombre in lista_urls:
    nombres.append(string.capwords(nombre.split('.es/')[1].split('/')[0].replace('-',' ')))
nombres
```

#### Creamos el DataFrame con las urls

```python
df_urls = pd.DataFrame()
df_urls['Nombre'] = nombres
df_urls['Urls'] = lista_urls
df_urls
```

```python
df_urls.to_excel('../data/Urls_Rend_Bajadas.xlsx')
```
