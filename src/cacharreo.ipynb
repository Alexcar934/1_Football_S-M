```python
import pandas as pd
```


1.	¿Cuál es el club que ha conseguido un mayor aumento en el valor de mercado de sus jugadores?

2.	¿Cuál es el club que ha logrado aumentar el valor de mercado de un mayor número de jugadores?


```python
df_conclusion1 = pd.DataFrame(df_market.groupby(['Club_Valor', 'Temporada'])['Diferencia_Valor']\
                              .sum()).reset_index().sort_values(by='Diferencia_Valor',ascending=False)
df_conclusion1[df_conclusion1['Temporada'] == 2022].head(20)
```

```python
df_market2 = df_market[['Nombre','Club_Valor','Diferencia_Valor','Temporada']]
```

```python
df_count = pd.DataFrame(df_market2[(df_market2['Diferencia_Valor']>0)]\
    .groupby(['Club_Valor','Temporada'])['Nombre'].count()).reset_index().sort_values(by='Nombre',ascending=False)
```

```python
df_sum = pd.DataFrame(df_market2[(df_market2['Diferencia_Valor']>0)]\
    .groupby(['Club_Valor','Temporada'])['Diferencia_Valor'].sum()).reset_index().sort_values(by='Diferencia_Valor',ascending=False)
```

```python
df_count.merge(df_sum, how='inner', on='Club_Valor')
```
